```php
$samples = [
    [3, 4, 50.5],
    [1, 5, 24.7],
    [4, 4, 62.0],
    [3, 2, 31.1],
];

$labels = ['married', 'divorced', 'married', 'divorced'];
```

```php
$dataset = new Labeled($samples, $labels);

```

```php
$estimator = new KNearestNeighbors(3);
```

```php
$estimator->train($dataset);
```

```php
var_dump($estimator->trained());
```
